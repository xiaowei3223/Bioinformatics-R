Like clustering, classification is also about categorizing data instances, but in this case, the categories are known and are termed as class labels. Thus, it aims at identifying the category that a new data point belongs to. It uses a dataset where the class labels are known to find the pattern. Classification is an instance of supervised learning where the learning algorithm takes a known set of input data and corresponding responses called class labels and builds a predictor model that generates reasonable predictions for the class labels in the unknown data. To illustrate, we use Breast Cancer Wisconsin Diagnostic dataset from http://aechive.ics.uci.edu/ml.


与分类一样，分类器也涉及对数据实例进行分类，但在本例中，类别是已知的，并被称为类标签。因此，它的目的是识别新数据点所属的类别。它使用已知类标签的数据集来查找模式。分类是监督学习的一个实例，学习算法取一组已知的输入数据和对应的响应类标签，建立预测模型，对未知数据中的类标签生成合理的预测。

There are some do's and don'ts to keep in mind while learning a classifier. You need to make sure that you have enough data to learn the model. Learning with smaller datasets will not allow the model to learn the pattern in an unbiased manner and again, you will end up with an inaccurate classification. Furthermore, the preprocessing steps (such as normalization) for the training and test data should be the same. Another important thing that one should take care of is to keep the training and test data distinct. Learning on the entire data and then using a part of this data for testing will lead to a phenomenon called overfitting. It is always recommended that you take a look at it manually and understand the question that you need to answer via your classifier.

在学习分类器时，要记住一些应该做和不应该做的事情。您需要确保有足够的数据来学习模型。使用更小的数据集进行学习将不允许模型以一种无偏见的方式学习模式，否则，您将得到一个不准确的分类。此外，训练和测试数据的预处理步骤(如标准化)应该是相同的。另一件需要注意的重要事情是保持训练和测试数据的独立性。对整个数据进行学习，然后使用其中一部分数据进行测试，将导致一种称为过度拟合的现象。建议您手动查看它，并理解需要通过分类器回答的问题.

There are several methods of classification. In this recipe, we will talk about some of these methods. We will discuss linear discriminant analysis (LDA), decision tree (DT), and support vector machine (SVM).

有几种分类方法。在本教程中，我们将讨论其中的一些方法。我们将讨论线性判别分析(LDA)、决策树(DT)和支持向量机(SVM)。

1.First, load the following MASS library as it has some of the classification functions:

    加载MASS包，里面包含一些分类器函数：

```r
library(MASS)
```

2. load the data,remove ID

```r
wbcd <- read.csv("wisc_bc_data.csv",stringsAsFactors = FALSE)#导入数据
head(wbcd)#查看数据
str(wbcd)
wbcd <- wbcd[-1] #剔除ID列
table(wbcd$Diagnosis)#统计良性肿块还是恶性肿块案例数
wbcd$Diagnosis <- factor(wbcd$Diagnosis,levels = c("B","M"),labels = c("Benign","Malignant")) ##将目标属性编码为因子类型，使用labels参数对B值和M值给出含有更多信息的标签
```

3. Randomly sample 380 data points for the training and the remaining 189 for the test set as follows-ensure that these two datasets do not overlap and are not biased towards any specific tumor type (random sampling):

 随机抽取380个数据点用于训练，其余189个数据点用于测试集，如下所示——确保这两个数据集不重叠，不偏向于任何特定的肿瘤类型(随机抽样):

```r
train_row <- sample(1:569, 380)
str(train_row)
class(train_row)
train <- wbcd[train_row,] # use sampled indexes to extract training data
test <- wbcd[-train_row,] # test set is select by selecting all the other data points
```

4. For the training data, retain the class labels, which are the tumor columns here,and remove this information from the test data. However, store this information
for comparison purposes:

 对于训练数据，保留类标签，也就是这里的肿瘤列，并从测试数据中删除这些信息。然而，储存这些信息是为了进行比较:

```r
testClass <- test$Diagnosis
test$Diagnosis <- NULL
```

##### 5. Now, try the linear discriminate analysis classifier, as follows, to get the classifier model:

 现在，尝试线性判别分析分类器，得到分类器模型如下:

```r
myLD <- lda(Diagnosis ~ .,train)
```

6. Test this classifier to predict the labels on your test set, as follows:

 这个分类器来预测测试集中的标签：

```r
testRes_lda <- predict(myLD, test)
```

7. To check the number of correct and incorrect predictions, simply compare the predicted classes with the testClass object, which was created in step 4, as follows:

 要检查正确和错误预测的数量，只需将预测的类与在步骤4中创建的testClass对象进行比较，如下所示：

```r
sum(testRes_lda$class == testClass) # correct prediction
```

```r
sum(testRes_lda$class != testClass) # incorrect prediction
```

```r
library(gmodels)
CrossTable(x=testRes_lda$class,y=testClass,
           prop.chisq =FALSE )
```

8. Now, try another simple classifier called DT. For this, you need the rpart package:

 尝试另一个简单的分类器DT（决策树），需要rpart包：

```r
library(rpart)
```

9. Create the decision tree based on your training data, as follows：

 根据您的培训数据创建决策树：

```r
myDT <- rpart(Diagnosis~ ., data = train, control = rpart.control(minsplit = 10))
```

10. Plot your tree by typing the following commands, as shown in the next diagram：

 通过输入以下命令来绘制树：

```r
plot(myDT）
text(myDT, use.n=T)
```

11. Now, test the decision tree classifier on your test data using the following prediction function：

 使用以下预测函数在测试数据上测试决策树分类器：

```r
testRes_dt <- predict(myDT, newdata= test)
```

12. Take a look at the species that each data instance is put in by the predicted classifier,as follows (1 if predicted in the class, else 0):
 
  查看由预测分类器放入的每个数据实例的种类，如下所示(如果在类中预测，则为1，否则为0)：

```r
classes <- round(testRes_dt)
```

```r
head(classes)
```

13. Finally, you'll work with SVMs. To be able to use them, you need another R package named e1071 as follows:

 SVM，需要e1071包：

```r
library(e1071)
```

14. Create the svm classifier from the training data as follows：

 从训练数据中创建svm分类器，如下所示：

```r
mySVM <- svm(Diagnosis ~ ., data = train)
```

15. Then, use your classifier, the model (mySVM object) learned to predict for the test data. You will see the predicted labels for each instance as follows:

 然后，使用svm分类器，模型(mySVM对象)学会了预测测试数据。您将看到每个实例的预测标签如下所示：

```r
testRes_svm <- predict(mySVM, test)
```

```r
testRes_svm
```
